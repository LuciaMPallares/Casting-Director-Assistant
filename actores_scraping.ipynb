```python
import pandas as pd
import numpy as np
import requests as req
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
from IPython.display import Image 
```

```python
path= ChromeDriverManager().install()
driver= webdriver.Chrome(path)
```

```python
url='https://www.imdb.com/list/ls044134118/'
driver.get(url)
```

```python
nombres=[]
urls=[]
fotos=[]

for i in range(3):
    paginas= driver.find_elements('xpath', '//h3[@class="lister-item-header"]/a')
    for nombre in paginas:
        nombres.append(nombre.text)
        
    direcciones= driver.find_elements('xpath', '//h3[@class="lister-item-header"]/a')
    for direccion in direcciones:
        urls.append(direccion.get_attribute('href'))
        
    imagenes= driver.find_elements('xpath', '//*[@id="main"]/div/div[2]//div[1]//a/img')
    for imagen in imagenes:
        fotos.append(imagen.get_attribute('src'))
    
    try:
        driver.find_element('xpath', '//a[@class="flat-button lister-page-next next-page"]').click()
        time.sleep(7)
    except:
        break
    
```

```python
columnas= ['actores', 'urls', 'fotos']
links=pd.DataFrame(list(zip(nombres, urls, fotos)), columns=columnas)
links.head()
```

```python
urls2=[]
actores=[]
premios=[]
generos1=[]
generos2=[]

for url in urls:
    driver.get(url)
    urls2.append(url)
    
    nombres= driver.find_element('xpath', '//h1[@data-testid="hero__pageTitle"]/span[1]').text
    actores.append(nombres)
    
    try:
        nominaciones= driver.find_element('xpath', '//li[@data-testid="award_information"]/div').text
    except:
        nominaciones=''
    premios.append(nominaciones)
    
    driver.find_element('xpath', '//button[@data-testid="hero-subnav-bar-all-topics-button"]').click()
    time.sleep(3)
    driver.find_element('xpath', '//a[text()="por género"]').click()
    time.sleep(3)
    
    try:
        cat1= driver.find_element('xpath', '//div[@class="faceter-fieldset genres"]/fieldset/span[1]').text
    except:
        cat1=''
        
    try:
        cat2= driver.find_element('xpath', '//div[@class="faceter-fieldset genres"]/fieldset/span[2]').text
    except:
        cat2=''
    generos1.append(cat1)   
    generos2.append(cat2)
    
        
```
