```python
import pandas as pd
import numpy as np
import requests as req
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
from IPython.display import Image 
```

```python
path= ChromeDriverManager().install()
driver= webdriver.Chrome(path)
```

```python
url='https://www.imdb.com/list/ls041877660/'
driver.get(url)
```

```python
actrices=[]      #coger el path que engloba los 100 elementos de cada pagina
urls=[]
fotos=[]
for i in range(4):
    paginas= driver.find_elements('xpath', '//h3[@class="lister-item-header"]/a')
    
    for nombre in paginas:
        actrices.append(nombre.text)
        
    direcciones= driver.find_elements('xpath', '//h3[@class="lister-item-header"]/a')
    for direccion in direcciones:
        urls.append(direccion.get_attribute('href'))
        
    imagenes= driver.find_elements('xpath', '//*[@id="main"]/div/div[2]/div[3]/div/div[1]/a/img')
    for imagen in imagenes:                   
        fotos.append(imagen.get_attribute('src'))  #hay actrices sin foto pero si tienen url es una foto avatar
    
    try:
        driver.find_element('xpath', '//a[@class="flat-button lister-page-next next-page"]').click()
        time.sleep(7)
    except:
        break
    
```

```python
columnas= ['actrices', 'urls', 'fotos']
links=pd.DataFrame(list(zip(actrices, urls, fotos)), columns=columnas)
links.head()
```

```python
links.to_csv(r'C:\Users\USER\Documents\Lucía\Ironhack\Bootcamp\Proyecto_final\CSV\links.csv', index=False)
```

```python
nombres=[]    #aqui los paths son de solo la actriz en cuestion
urls2=[]
premios=[]
generos1=[]
generos2=[]

for url in urls:  #pasa por cada link y abrelo
    driver.get(url)
    urls2.append(url) #guarda el url
    
    nombre= driver.find_element('xpath', '//h1[@data-testid="hero__pageTitle"]/span[1]').text
    nombres.append(nombre) #encuentra y guarda el nombre en texto                        
                              
    
    try:
        premio=driver.find_element('xpath', '//li[@data-testid="award_information"]/div').text
    except:     #encuentra y guarda el nombre excepto si está vacío
        premio= ''    
    premios.append(premio)  #guarda premios
        
    driver.find_element('xpath', '//button[@data-testid="hero-subnav-bar-all-topics-button"]').click()
    time.sleep(2)   #abre ventana
    driver.find_element('xpath', '//a[text()="por género"]').click()
    time.sleep(2)  #abre ventana generos
    
    try:
        cat1=driver.find_element('xpath', '//div[@class="faceter-fieldset genres"]/fieldset/span[1]').text
    except:     
        cat1= ''
    
    try:
        cat2=driver.find_element('xpath', '//div[@class="faceter-fieldset genres"]/fieldset/span[2]').text
    except:
        cat2= ''
    generos1.append(cat1)  #guarda el primero y el segundo
    generos2.append(cat2)
    
    
    


        
```

```python
columnas= ['nombres', 'premios', 'generos1', 'generos2', 'urls']
df=pd.DataFrame(list(zip(nombres, premios, generos1, generos2, urls2)), columns=columnas)
df.tail()
```
