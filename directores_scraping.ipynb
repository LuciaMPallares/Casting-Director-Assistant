```python
import pandas as pd
import numpy as np
import requests as req
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
from IPython.display import Image 
```

```python
path= ChromeDriverManager().install()
driver= webdriver.Chrome(path)
```

```python
url='https://www.imdb.com/list/ls552103190/'
driver.get(url)
```

```python
directores=[]
urls=[]
fotos=[]

for i in range(2):
    
    nombres= driver.find_elements('xpath', '//h3[@class="lister-item-header"]/a')
    for nombre in nombres:
        directores.append(nombre.text)
    
    direcciones= driver.find_elements('xpath', '//h3[@class="lister-item-header"]/a')
    for direccion in direcciones:
        urls.append(direccion.get_attribute('href'))
    
    imagenes= driver.find_elements('xpath', '//div[@class="lister-item-image"]/a/img')
    for imagen in imagenes:
        fotos.append(imagen.get_attribute('src'))
    
    try:
        driver.find_element('xpath', '//a[@class="flat-button lister-page-next next-page"]').click()
        time.sleep(7)
    except:
        break
```

```python
columnas=['directores', 'fotos', 'urls']
linkdirectores= pd.DataFrame(list(zip(directores,urls,fotos)), columns=columnas)
linkdirectores.tail()
```

```python
nombres=[]
urls2=[]
premios=[]
generos1=[]
generos2=[]

for url in urls:
    driver.get(url)
    urls2.append(url)
    
    direct= driver.find_element('xpath', '//h1[@data-testid="hero__pageTitle"]/span[1]').text
    nombres.append(direct)
    
    try:
        nominaciones= driver.find_element('xpath', '//div[@class="ipc-metadata-list-item__content-container"]/ul/li/label').text
    except:
        nominaciones=''
    premios.append(nominaciones)
        
    driver.find_element('xpath', '//button[@data-testid="hero-subnav-bar-all-topics-button"]').click()
    time.sleep(3)
    driver.find_element('xpath', '//a[text()="por género"]').click()
    time.sleep(3)
        
    try:
        cat1= driver.find_element('xpath', '//div[@class="faceter-fieldset genres"]/fieldset/span[1]').text
    except:
        cat1=''
    
    try:
        cat2= driver.find_element('xpath', '//div[@class="faceter-fieldset genres"]/fieldset/span[2]').text
    except:
        cat2=''
        
    generos1.append(cat1)
    generos2.append(cat2)
    
```

```python
columnas= ['directores', 'premios', 'generos1', 'generos2', 'urls2']
generosdirectores= pd.DataFrame(list(zip(nombres,premios,generos1,generos2,urls2)), columns=columnas)
generosdirectores.head(20)        
```
